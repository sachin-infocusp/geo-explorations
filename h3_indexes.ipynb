```python
import h3
import numpy as np
import folium
from global_land_mask import globe
import pandas as pd

all_h3_indexes = set()

for lat in np.arange(-90.0, 90.0, 0.2):
    for lng in np.arange(-180.0, 180.0, 0.2):
        
        i = h3.geo_to_h3(lat, lng, resolution=2)
        all_h3_indexes.add(i)

print(len(all_h3_indexes))

```

```python
final_points = []


map = folium.Map(location=[41.492537, -99.901813], zoom_start=5)

for h3_point in all_h3_indexes:
    point = h3.h3_to_geo(h3_point)

    if globe.is_land(point[0], point[1]):

        folium.Circle(point).add_to(map)

        final_points.append(point)

map
```

```python
env_df = pd.DataFrame(final_points, columns=['Latitude','Longitude'])

env_df['Env'] = env_df.apply(
    lambda r: str('Location_' + str(r['Latitude'])[:5]) + '_' + str(r['Longitude'])[:5],axis=1)

env_df
```
